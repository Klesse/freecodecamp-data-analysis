```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/medical_examination.csv')
df
```

```python
df['overweight'] = ((df['weight']/((df['height']/100)**2)) > 25).astype(int)
```

```python
df[['gluc','cholesterol']] = (df[['gluc','cholesterol']] > 1).astype(int)
```

```python
df_cat = pd.melt(df,['cholesterol','gluc','smoke','alco','active','overweight'])
df_cat
```

```python
df_cat = pd.melt(df,id_vars=['cardio'],value_vars=['active','alco','cholesterol','gluc','overweight','smoke'])
df_cat
```

```python
sns.catplot(data = df_cat, kind='count',  x='variable', hue='value', col='cardio').set(ylabel = 'total')
```

```python
df
```

```python
df_heat = df[
    (df['ap_lo'] <= df['ap_hi']) &
    (df['height'] >= df['height'].quantile(0.025)) &
    (df['height'] <= df['height'].quantile(0.975)) &
    (df['weight'] >= df['weight'].quantile(0.025)) &
    (df['weight'] <= df['weight'].quantile(0.975))
]
```

```python
corr = df_heat.corr()
corr
```

```python
mask = np.triu(corr)
mask
```

```python
fig, ax =  plt.subplots(figsize=(14,6))
ax = sns.heatmap(corr, mask=mask, annot=True, fmt='.1f',cmap='mako',square=True)
```
